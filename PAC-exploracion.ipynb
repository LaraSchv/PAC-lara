```python
def cortar_fecha(fecha_y_hora):
  return fecha_y_hora.split(' ')[0]

df['Marca temporal'] = df['Marca temporal'].apply(cortar_fecha)
df['Marca temporal'] = pd.to_datetime(df['Marca temporal'],format='%Y/%m/%d')
```

Paso 1: parsear la columna de marca temporal para que sea una fecha y hora (una columna cada cosa).

Paso 2: Separar las fechas por semanas; crear un nuevo atributo que identifique las entregas.

Para el encuentro del jueves 12 queremos modificar la columna de fecha para poder leer con datetime la fecha, hora, etc; y luego crear los atributos y columnas mencionados en los pasos anteriores.

```python
#ya tenemos las fechas convertidas a datetime

#Para identificar la fecha de entrega correspondiente a un pedido, hallamos la fecha del primer s√°bado posterior.

def sabado_posterior(fecha):
  sabado = fecha
  while(sabado.weekday() != 5):
    sabado = sabado + timedelta(days = 1)
  return sabado

sabado_posterior(df['Marca temporal'].loc[0])
```

```python
df['Fechas entrega'] = df['Marca temporal'].apply(sabado_posterior) #estas nuevas fechas son tipo datetime
df['Fechas entrega'].unique()
#YA TENEMOS LAS QUINCENAS
```
